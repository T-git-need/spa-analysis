```python
# 正确的清除方式
import sys

# 只删除你的模块，保留系统模块
modules_to_remove = []
for k in sys.modules.keys():
    if k in ['spa_analyzer', 'comparison_system', 'main', 'spa_analyzer', 'comparison_system']:
        modules_to_remove.append(k)

for m in modules_to_remove:
    if m in sys.modules:
        del sys.modules[m]

print(f"已清除 {len(modules_to_remove)} 个模块")
```
