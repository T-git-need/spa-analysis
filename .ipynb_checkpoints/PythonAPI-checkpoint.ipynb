```python
import sys
import os
import glob
import shutil
```

```python
# 基础路径
openpose_dir = r'D:\Jupyter.project\spa-analysis\openpose-prosperity'
target_dir = os.path.join(openpose_dir, 'build_GPU', 'python', 'openpose', 'Release')

# 确保目标文件夹存在
os.makedirs(target_dir, exist_ok=True)

# 需要复制的源文件夹
source_dirs = [
    os.path.join(openpose_dir, 'build_GPU', 'bin'),
    os.path.join(openpose_dir, 'build_GPU', 'x64', 'Release'),
    os.path.join(openpose_dir, '3rdparty', 'windows', 'caffe', 'bin'),
    os.path.join(openpose_dir, '3rdparty', 'windows', 'caffe_cpu', 'bin'),
    os.path.join(openpose_dir, '3rdparty', 'windows', 'freeglut', 'bin'),
    os.path.join(openpose_dir, '3rdparty', 'windows', 'opencv', 'x64', 'vc15', 'bin'),
]

copied_count = 0
for src_dir in source_dirs:
    if os.path.exists(src_dir):
        dll_files = glob.glob(os.path.join(src_dir, '*.dll'))
        print(f"📁 {src_dir}: 找到 {len(dll_files)} 个 DLL")
        
        for dll in dll_files:
            target_file = os.path.join(target_dir, os.path.basename(dll))
            if not os.path.exists(target_file):
                shutil.copy2(dll, target_dir)
                copied_count += 1
                print(f"  ✅ 复制: {os.path.basename(dll)}")
    else:
        print(f"❌ 不存在: {src_dir}")

print(f"\n{'='*50}")
print(f"总共复制了 {copied_count} 个 DLL 文件到:")
print(f"{target_dir}")
print(f"{'='*50}")

# 列出目标文件夹中的所有文件
print(f"\n📂 目标文件夹现在包含:")
files = os.listdir(target_dir)
for f in sorted(files):
    size = os.path.getsize(os.path.join(target_dir, f))
    print(f"  {f} ({size/1024:.1f} KB)")
```

```python
# 只需要添加这一个路径
openpose_dir = r'D:\Jupyter.project\spa-analysis\openpose-prosperity'
python_path = os.path.join(openpose_dir, 'build_GPU', 'python', 'openpose', 'Release')

# 添加到 Python 路径
if python_path not in sys.path:
    sys.path.insert(0, python_path)

# 添加 DLL 路径
os.environ['PATH'] = python_path + ';' + os.environ['PATH']

print(f"Python 路径: {python_path}")
print(f"路径存在: {os.path.exists(python_path)}")

# 尝试导入
try:
    import pyopenpose as op
    print("\n✅✅✅ pyopenpose 导入成功！✅✅✅")
except Exception as e:
    print(f"\n❌ 仍然失败: {e}")
```

```python
# 测试 OpenPose 参数设置
params = {
    "model_folder": os.path.join(openpose_dir, "models"),
    "hand": False,
    "face": False
}

# 创建 OpenPose 包装器
opWrapper = op.WrapperPython()
opWrapper.configure(params)
opWrapper.start()

print("✅ OpenPose 初始化成功！")
```
