```python
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt
```

```python
# 初始化 MediaPipe 面部网格
mp_face_mesh = mp.solutions.face_mesh
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles

face_mesh = mp_face_mesh.FaceMesh(
    static_image_mode=True,
    max_num_faces=5,  # 最多检测 5 张脸
    refine_landmarks=True,  # 细化眼睛和嘴唇
    min_detection_confidence=0.5
)
```

```python
# 读取图片
image_path = r"D:\facetest.png"
image = cv2.imread(image_path)

if image is not None:
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # 检测面部
    results = face_mesh.process(image_rgb)
    
    if results.multi_face_landmarks:
        print(f"✅ 检测到 {len(results.multi_face_landmarks)} 张人脸")
        
        # 绘制结果
        annotated_image = image_rgb.copy()
        for idx, face_landmarks in enumerate(results.multi_face_landmarks):
            print(f"\n人脸 {idx+1}: {len(face_landmarks.landmark)} 个关键点")
            
            # 绘制面部网格
            mp_drawing.draw_landmarks(
                image=annotated_image,
                landmark_list=face_landmarks,
                connections=mp_face_mesh.FACEMESH_TESSELATION,
                landmark_drawing_spec=None,
                connection_drawing_spec=mp_drawing_styles.get_default_face_mesh_tesselation_style()
            )
            
            # 绘制轮廓
            mp_drawing.draw_landmarks(
                image=annotated_image,
                landmark_list=face_landmarks,
                connections=mp_face_mesh.FACEMESH_CONTOURS,
                landmark_drawing_spec=None,
                connection_drawing_spec=mp_drawing_styles.get_default_face_mesh_contours_style()
            )
        
        # 显示结果
        plt.figure(figsize=(14, 10))
        plt.imshow(annotated_image)
        plt.axis('off')
        plt.title(f"MediaPipe 面部关键点检测 - {len(results.multi_face_landmarks)} 张人脸")
        plt.show()
    else:
        print("❌ 未检测到人脸")
```

```python
face_mesh.close()
```
