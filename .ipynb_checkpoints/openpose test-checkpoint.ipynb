```python
import sys
import os
import cv2

# ========== 路径设置 ==========
openpose_dir = r'D:\Jupyter.project\spa-analysis\openpose-prosperity'

# 添加 Python 接口路径
sys.path.append(os.path.join(openpose_dir, 'build_GPU', 'python', 'openpose', 'Release'))

# 添加 DLL 路径
os.environ['PATH'] = os.environ['PATH'] + ';' + os.path.join(openpose_dir, 'build_GPU', 'python', 'openpose', 'Release')

# 强制使用 CUDA 11.0
cuda_path = r'C:\Program Files\NVIDIA GPU Computing Toolkit\CUDA\v11.0'
os.environ['CUDA_PATH'] = cuda_path
os.environ['CUDA_HOME'] = cuda_path
os.environ['PATH'] = os.path.join(cuda_path, 'bin') + ';' + os.environ['PATH']



```

```python
import pyopenpose as op
# ========== 配置参数 ==========
params = {
    "model_folder": os.path.join(openpose_dir, "models"),
    "number_people_max": 1,  # 最多检测 1 个人
    "disable_blending": False,
    "net_resolution": "128x96",  # 低分辨率，避免崩溃
}

```

```python
# ========== 初始化 ==========
opWrapper = op.WrapperPython()
opWrapper.configure(params)
opWrapper.start()
print("✅ OpenPose 启动成功！")

```

```python
# ========== 处理图片 ==========
image_path = r"D:\test.png"
image = cv2.imread(image_path)

if image is not None:
    print(f"图片尺寸: {image.shape}")
    
    datum = op.Datum()
    datum.cvInputData = image
    
  
```
